```python
from experta import *

class HeartDiagnosis(Fact):
    pass

class HeartExpertSystem(KnowledgeEngine):

    def ask_yes_no(self, question):
        response = ""
        while response not in ["نعم", "لا"]:
            response = input(f"{question} (نعم/لا): ").strip()
        return response == "نعم"

    @DefFacts()
    def initial_facts(self):
        yield Fact(action="start")

    @Rule(Fact(action="start"))
    def start_diagnosis(self):
        self.declare(HeartDiagnosis(chest_pain=self.ask_yes_no("هل تعاني من ألم في الصدر؟")))


    @Rule(HeartDiagnosis(chest_pain=True))
    def check_pressing_pain(self):
        self.declare(HeartDiagnosis(pressing_pain=self.ask_yes_no("هل الألم ضاغط أو كأن شيء يعصر صدرك؟")))

    @Rule(HeartDiagnosis(pressing_pain=True))
    def check_radiation(self):
        self.declare(HeartDiagnosis(radiation=self.ask_yes_no("هل ينتشر الألم إلى الذراع الأيسر أو الفك أو الظهر؟")))


    @Rule(HeartDiagnosis(radiation=True))
    def check_sweating(self):
        self.declare(HeartDiagnosis(sweating=self.ask_yes_no("هل تعاني من تعرق شديد مع الألم؟")))

    @Rule(HeartDiagnosis( sweating=True))
    def heart_attack(self):
        print(" نقص تروية يؤدي لاحتشاء عضلة القلب")
        self.halt()

    @Rule(HeartDiagnosis(sweating=False))
    def check_exertion(self):
        self.declare(HeartDiagnosis(exertion=self.ask_yes_no("هل يزداد الألم مع الجهد و الراحة؟")))

    @Rule(HeartDiagnosis( exertion=True))
    def stable_angina(self):
        print("خناق صدري يؤلم بالجهدوراحة")
        self.halt()

    @Rule(HeartDiagnosis(exertion=False))
    def unstable_angina(self):
        print("خناق صدري مستر قد يشتد على جهد ويخف بالرحة")
        self.halt()


    @Rule(HeartDiagnosis(radiation=False))
    def check_duration(self):
        self.declare(HeartDiagnosis(long_duration=self.ask_yes_no("هل يستمر الألم أكثر من 20 دقيقة من الراحة؟")))

    @Rule(HeartDiagnosis(long_duration=True))
    def acute_coronary(self):
        print("متلازمة الشريان التاجي الحاد")
        self.halt()

    @Rule(HeartDiagnosis( long_duration=False))
    def angina_or_muscle(self):
        print("خناق صدري أو ألأم عضلي")
        self.halt()

    @Rule(HeartDiagnosis( pressing_pain=False))
    def check_sharp_pain(self):
        self.declare(HeartDiagnosis(sharp_pain=self.ask_yes_no("هل الألم حاد يزداد مع التنفس العميق؟")))

    @Rule(HeartDiagnosis(sharp_pain=True))
    def pericarditis(self):
        print("التهاب غشاء القلب أو مشكلة رئوية")
        self.halt()

    @Rule(HeartDiagnosis(sharp_pain=False))
    def non_cardiac(self):
        print("ألم صدري غير قلبي")
        self.halt()


    @Rule(HeartDiagnosis(chest_pain=False))
    def check_dyspnea(self):
        self.declare(HeartDiagnosis(dyspnea=self.ask_yes_no("هل تعاني من ضيق في التنفس؟")))


    @Rule(HeartDiagnosis( dyspnea=True))
    def check_exertion_dyspnea(self):
        self.declare(HeartDiagnosis(exertion_dyspnea=self.ask_yes_no("هل يحدث ضيق التنفس مع مجهود بسيط؟")))

    @Rule(HeartDiagnosis(exertion_dyspnea=True))
    def check_swelling(self):
        self.declare(HeartDiagnosis(swelling=self.ask_yes_no("هل تعاني من تورم في القدمين أو الساقين؟")))

    @Rule(HeartDiagnosis( swelling=True))
    def heart_failure(self):
        print("قصور القلب")
        self.halt()

    @Rule(HeartDiagnosis(swelling=False))
    def check_lying(self):
        self.declare(HeartDiagnosis(lying=self.ask_yes_no("هل يزداد ضيق التنفس عند الاستلقاء؟")))

    @Rule(HeartDiagnosis( lying=True))
    def left_heart_failure(self):
        print(" قصور القلب الأيسر")
        self.halt()

    @Rule(HeartDiagnosis( lying=False))
    def dyspnea_evaluation(self):
        print("ضيق تنفس")
        self.halt()

    @Rule(HeartDiagnosis( exertion_dyspnea=False))
    def mild_dyspnea(self):
        print("ضيق تنفس خفيف")
        self.halt()

    @Rule(HeartDiagnosis( dyspnea=False))
    def check_palpitations(self):
        self.declare(HeartDiagnosis(palpitations=self.ask_yes_no("هل تشعر بخفقان أو عدم انتظام في ضربات القلب؟")))

    @Rule(HeartDiagnosis( palpitations=True))
    def check_fainting(self):
        self.declare(HeartDiagnosis(fainting=self.ask_yes_no("هل سبق وفقدت الوعي أو شعرت بدوخة شديدة؟")))

    @Rule(HeartDiagnosis( fainting=True))
    def dangerous_arrhythmia(self):
        print("اضطراب نظم القلب")
        self.halt()

    @Rule(HeartDiagnosis( fainting=False))
    def check_frequency(self):
        self.declare(HeartDiagnosis(frequent=self.ask_yes_no("هل الخفقان متكرر (أكثر من مرة أسبوعياً)؟")))

    @Rule(HeartDiagnosis( frequent=True))
    def arrhythmia(self):
        print("  اضطراب نظم القلب")
        self.halt()

    @Rule(HeartDiagnosis( frequent=False))
    def benign_palpitations(self):
        print("خفقان طبيعي غالبا بصحة جيدة")
        self.halt()

    @Rule(HeartDiagnosis(palpitations=False))
    def check_fatigue(self):
        self.declare(HeartDiagnosis(fatigue=self.ask_yes_no("هل تعاني من إرهاق شديد غير مبرر؟")))

    @Rule(HeartDiagnosis( fatigue=True))
    def possible_heart(self):
        print("قد يكون مرتبط بالقلب أو أعضاء أخرى")
        self.halt()

    @Rule(HeartDiagnosis( fatigue=False))
    def no_symptoms(self):
        print("لاتوجد أعراض قلبية واضحة")
        self.halt()


if __name__ == "__main__":

    print(" نظام خبير لتشخيص أمراض القلب")


    start = input("هل ترغب في البدء؟ (نعم/لا): ").strip()

    if start == "نعم":
        engine = HeartExpertSystem()
        engine.reset()
        engine.run()
    else:
        print("ربما في وقت لاحق!")

    print("\nشكراً لاستخدامك نظام تشخيص أمراض القلب!")
```
